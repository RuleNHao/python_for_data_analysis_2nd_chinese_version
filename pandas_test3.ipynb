```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = 'all'
```

```python
# 使用数据a = pd.DataFrame的列或多个列当做行索引
a = pd.DataFrame(np.random.rand(5,6))
a['rank'] = list('abcde')
a = a.set_index(['rank'])
a
a = a.reset_index()
a
```

```python
# 使用数据连接方法
# pd.concat([df1, df2, ...], axis=0/1, join='inner'/'outer')
a = pd.DataFrame(np.random.rand(3,4), columns=list('abcd'))
a
b = pd.DataFrame(np.random.rand(2,3), columns=list('abd'))
b
pd.concat([a, b], ignore_index=True)
pd.concat([a, b]).reset_index(drop=True)
c = pd.concat([a, b], axis=1)
c
for columns in a.columns:
    if columns != 'c':
        c[columns+'_sum'] = c[columns].sum(axis=1)
    else:
        c[columns+'_sum'] = c[columns]
    del c[columns]
map = {}
for key, value in zip(c.columns, a.columns):
    map[key] = value

c = c.rename(columns=map);
c
```

```python
# DataFrame的stack/unstack可以将数据转换为multiIndex的Series
a = pd.DataFrame(np.random.rand(4,5))
a
b = a.stack()
c = a.unstack()
b.index
c
```
