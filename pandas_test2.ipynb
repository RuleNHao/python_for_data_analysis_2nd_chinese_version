```python
# pandas中对缺失值的处理
a = pd.DataFrame(np.random.randint(1, 10, (4,5)))
a.iloc[[2,3],[1]] = np.nan
a.iloc[:,4] = np.nan
a.isnull()
b = a.fillna(100)
b
a.dropna(axis=1,how='all')
a.dropna(axis=1, thresh=4)
a.fillna(value=a[3].mean(), axis=1, limit=2)
a[1] = a[1].fillna(value=0)
a[4] = a[4].fillna(value=a[1].mean())
a
```

```python
# 替换值和重命名轴索引
a = pd.DataFrame(np.linspace(1,10,num=10).reshape(2,5))
# 创建字典映射
transform = {1: 100, 2:200, 10:1000}
a.replace(transform)
a.replace([1, 2], [100, 200])
a[5] = a.sum(axis=1)
# 创建轴标签字典
index1 = np.array(a.index)
index2 = np.array([1,2])
transf = {}
for key, value in zip(index1, index2):
    transf[key] = value
transf
#a.index = a.index.map(transf)
b = a.rename(index=transf)
a.rename(index=transf, columns=transf)
a.reindex(index=[1,2])
# 注意！！！
# rename和reindex的区别！！！
# rename是重新更改数据的索引，reindex是找到新的索引位置，和切片功能很相似！！！！
```

```python
# 计算指标矩阵，分类任务中通常将单个label转换为相应的矩阵
# 下例为创建随机一个feature~label的数据框架，将label转换为指标矩阵
np.random.seed(1234)
source_data = {'feature1': np.random.rand(10), 'feature2': np.random.rand(10), 'label': np.random.randint(0,10,10)}
data = pd.DataFrame(source_data)
data
data_label = pd.get_dummies(data['label'], prefix='label')
data = data.join(data_label)
del data['label']
data
```
