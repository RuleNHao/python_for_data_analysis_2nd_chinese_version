```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
#from IPython.core.interactiveshell import InteractiveShell
#InteractiveShell.ast_node_interactivity = 'all'
```

```python
a = np.arange(10)
plt.plot(a);
```

```python
# 实用的子图功能创建
# plt.subplots(nrow, ncols, sharex, sharey, figsize=(,))
# 当使用sharex和sharey时,调节xlim将会是对整个axes调节
a = np.random.rand(50)
b = np.random.rand(100)
c = np.random.rand(200)
fig, axes = plt.subplots(2,3, figsize=(10,8), sharex=True, sharey=True)
axes[0,0].plot(a.cumsum())
axes[0,1].plot(b)
axes[1,0].plot(c.cumsum())
axes[0,0].set_xlim(0,50)
axes[0,0].set_ylim(0,20)


# 调整subplot之间的间距
# subplots_adjust(left=None, bottom=None, right=None, top=None, wspace=None, hspace=None)
# 一般用来调整wspace和hsapce,调整子图的间距
plt.subplots_adjust(wspace=0, hspace=0)
```

```python
# 对图像颜色,标记,线型的处理
# 使用plot?获得相应的参数
plt.plot?
```

```python
# 展示部分图表样例
# 建议不使用plt高级API和'go--'类似的混合格式,而是把每个细节展示出来,熟练后再简写
# 常用的图形参数
# color 线条颜色
# label 给线条标记,后续使用legend显示出来
# drawstyle 线条拟合方式
# linestyle 线条形状
# linewidth 线条宽度
# marker 标记数据点的样式
# markersize 标记的大小
# markerfacecolor 标记填充颜色
# fillstyle 标记填充颜色的位置'full'/'right'/'left'/'top'/'bottom'
# markeredgecolor 标记边缘填充颜色
# markeredgewidth 标记边缘宽度
# **other_krg
x = np.arange(10)
y = np.random.rand(10)
data = pd.DataFrame({'x':x, 'y':y}, dtype=np.float32)
fig, axes = plt.subplots(2, 2, figsize=(10,8), sharex=True, sharey=True)
axes[0,0].plot(x, y, label='line1', color='m', linestyle='-', linewidth=2, marker='D', markersize=14, markerfacecolor='b',
              markeredgecolor='g', drawstyle='steps', fillstyle='left')
axes[0,1].plot(x, y, label='line2', color='green', linestyle=':', linewidth=2, marker='^', markersize=12,
              drawstyle='steps-pre', fillstyle='right')
axes[1,0].plot(x, y, label='line3', color='blue', linestyle='-.', linewidth=2, marker='s', markersize=12,
              drawstyle='steps-post', fillstyle='top', markerfacecolor='black')
axes[1,1].plot(x, y, label='line4', color='black', linestyle='--', linewidth=2, marker='*', markersize=12,
              drawstyle='steps-mid', fillstyle='bottom')
for i, j in zip([0,0,1,1], [0,1,0,1]):
    axes[i,j].legend()
plt.subplots_adjust(wspace=0.1, hspace=0.1)
```

```python
# 针对plt和axes两种方法,plt是过程方法,axes是对象方法
# 通常在处理多个子图中axes方法更为明确'
# 设置标题,轴标签,刻度,刻度标签
x = np.random.randn(1000)
fig, ax = plt.subplots(1, 1, figsize=(15,8))
ax.plot(x.cumsum(), label='Random cumsum', color='black', linestyle='-', marker='.', markerfacecolor='r')
ax.set_xlim(-1,1001)
ax.set_title('Random cumsum', color='b', fontsize=22)
ax.set_xlabel('Number', color='r', fontsize=16)
ax.set_ylabel('Location', c='r', fontsize=16, rotation=90)
ax.set_xticks([100, 200, 500, 800])
ax.set_xticklabels(['one', 'two', 'three', 'four'], rotation=30, fontsize='large', color='r')
ax.set_yticks
ax.set_yticklabels
ax.legend(loc='best', fontsize='xx-large');
```

```python
# 在axes上某些特定位置绘图,注释,其他图形
# ax.text
# ax.arrow
# ax.annotate
np.random.seed(1234)
x = np.random.randn(1000)
fig, ax = plt.subplots(1, 1, figsize=(15,8))
ax.plot(x.cumsum(), color='k', linestyle='-', marker='.', markerfacecolor='w')
ax.text(200, 30, '$E=MC^2$', color='g', fontsize='xx-large');

# 在图中添加图形,需要添加块对象(patch)
# 更多操作可以参考matplotlib官方库
patch1 = plt.Circle((400, 5), 5, color='g', alpha=0.3)
ax.add_patch(patch1)
```
